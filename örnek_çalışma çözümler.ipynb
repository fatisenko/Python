1. Bir spor müsabakasına katılan sporcuların skorlarının olduğu bir liste üzerinden, ikinci olan sporcuyu çıktı veren bir kod yazın.

   
   örnek liste = [3,4,2,6,5,6]
   
istenen çıktı = 5

```python
liste = [3,4,2,6,5,6]
ikinci = []
for i in liste:
  if i != max(liste):
    ikinci.append(i)
print(max(ikinci))    
```

5.

Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: !) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

5! = 5 × 4 × 3 × 2 × 1 = 120
Expected output:

120

```python
def faktoriyel(n):
  faktoriyel_ = 1
  for i in range(1,n+1):
    faktoriyel_ *= i
  return faktoriyel_
print(faktoriyel(5))  
```
